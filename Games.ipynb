## IMPORTANDO BIBLIOTECAS

```python
# Importando Lib Pandas
import pandas as pd

# Informar ao Matplotlib sobre os gráficos inline
%matplotlib inline

#Importando Lib Matplotlib
import matplotlib.pyplot as plt

#Importando lib Numpy
import numpy as np
from scipy import stats

#Importando lib Seaborn
import seaborn as sns
```

## CARREGANDO DATASET (ONLINE)

```python
# abrindo o dataset com a lib Pandas
df = pd.read_csv('https://query.data.world/s/ber34nftkhskriwesleohahiq2db2z')
```

## REFINANDO O DATASET

```python
dfRefined = df[['Pokemon','Trainer Region','Level','Type1','Type2','Nature','Pokeball']]
dfRefined
```

## CORRIGINDO O DATASET

```python
# Renomeando indice
dfRefined.rename(columns={'Trainer Region':'TrainerRegion'}, inplace = True)
dfRefined
```

## ORGANIZANDO O DADOS DO DATASET

```python
# Ordenando Dataset em forma crescente
df_ordenado = dfRefined.sort_values(by='Level')
df_ordenado
```

```python
#Ordenando em ordem decrescente
dfDesc = dfRefined.sort_values(by='Level', ascending=False)
dfDesc
```

## CRIANDO GRÁFICOS E RESULTADOS A PARTIR DO DATASET

**GRÁFICO DE BARRA - NÍVEIS DOS POKEMONS ENVOLVIDOS NO WONDER TRADE**

```python
p, ax = plt.subplots(figsize=(15, 4))

sns.set_color_codes("muted")
sns.barplot(x=df_ordenado.Pokemon.tail(60), y=df_ordenado.Level.tail(60), palette="muted")

ax.set(ylim=(0, 60),
       ylabel="Level",
       xlabel="Pokemon")

ax.set_xticklabels(ax.get_xticklabels(),rotation=65, horizontalalignment='right')

sns.despine(left=True, bottom=True)
```

```python
print('O MAIOR LEVEL É:') 
np.max(dfRefined.Level)
```

```python
print('O MENOR LEVEL É:') 
np.min(dfRefined.Level)
```

**GRÁFICO DE PIZZA - INDICE DA NATUREZA DOS POKEMONS ENVOLVIDOS NO WONDER TRADE**

```python
totalNature = list(dfRefined.Nature.drop_duplicates())

plt.subplots(figsize=(15, 15))
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"] 
plt.pie(dfRefined.Nature.value_counts(), labels=totalNature, colors=colors,
autopct='%1.0f%%', shadow=True, startangle=1000)
plt.title("Indice da natureza dos Pokemons envolvidos no Wonder Trade")
plt.show()
```
